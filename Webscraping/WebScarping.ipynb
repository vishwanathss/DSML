```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://www.amazon.in/OnePlus-Nebula-Blue-256GB-Storage/product-reviews/B07HGJKDQB/ref=cm_cr_dp_d_show_all_top?ie=UTF8&reviewerType=all_reviews'
```

```python
review_html=requests.get(url)
```

```python
review_html=review_html.text
```

```python
review_html
```

```python
soup=BeautifulSoup(review_html,'lxml')
```

```python
soup
```

```python
review_titles=soup.findAll('a',{'class':'a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold'})
```

```python
titles=[]
for title in review_titles:
    titles.append(title.text)
titles=pd.Series(titles)
```

```python
review_body=soup.findAll('span',{'class':'a-size-base review-text review-text-content'})
```

```python
body=[]
for line in review_body:
    body.append(line.text)
body=pd.Series(body)
```

```python
name=soup.findAll('div',{'class':'a-profile-content'})
```

```python
names=[]
for noun in name:
    names.append(noun.text)
names=pd.Series(names)
```

```python
review_data=pd.concat([names,titles,body],axis=1)
```

```python
review_data.columns=['Names','Titles','Comment']
```
